# Binary Math Hacks

How do you think we should find the difference of two binary numbers? The product? The quotient? What rules need to be followed for those operations?

To find the difference of two binary numbers, you need to subtract the second number from the first number, following the same rules as subtracting decimal numbers. Start with the rightmost digits and subtract them. If the result is negative, borrow 1 from the next digit to the left. Then, subtract the next pair of digits and repeat until you have subtracted all digits. The final result is the difference between the two numbers.To find the product of two binary numbers, you need to multiply the two numbers, following the same rules as multiplying decimal numbers. Start with the rightmost digit of the second number and multiply it by the first number. Then, shift the second number one digit to the left and repeat the process. Add the results of each multiplication to get the final product. To find the quotient of two binary numbers, you need to use long division, as you would with decimal numbers. Start with the leftmost digits of the dividend and divisor. Divide the dividend by the divisor, and write the quotient below the dividend. Multiply the quotient by the divisor, and subtract the result from the dividend. Bring down the next digit of the dividend and repeat the process until you have divided all digits. The final result is the quotient. The rules for binary arithmetic operations are the same as the rules for decimal arithmetic operations. The only difference is that in binary, there are only two digits (0 and 1), so the carry and borrow operations may differ slightly. However, the basic principles of addition, subtraction, multiplication, and division are the same.

```python
### Creating the buttons for all the Hacks
# Binary subtraction function
def binary_subtraction(binary1, binary2):
    decimal1 = int(binary1, 2)
    decimal2 = int(binary2, 2)
    decimal = decimal1 - decimal2
    binary = bin(decimal)[2:]
    return (binary, decimal)

# Binary multiplication function
def binary_multiplication(binary1, binary2):
    decimal1 = int(binary1, 2)
    decimal2 = int(binary2, 2)
    decimal = decimal1 * decimal2
    binary = bin(decimal)[2:]
    return (binary, decimal)

# Binary division function
def binary_division(binary1, binary2):
    decimal1 = int(binary1, 2)
    decimal2 = int(binary2, 2)
    if decimal2 == 0:
        return ("Undefined", "Undefined")
    decimal = decimal1 // decimal2
    binary = bin(decimal)[2:]
    return (binary, decimal)

# Take user input for binary numbers
binary1 = input("Enter binary number 1: ")
binary2 = input("Enter binary number 2: ")

# Perform binary subtraction
binary, decimal = binary_subtraction(binary1, binary2)
print("Subtraction Result:")
print("Binary:", binary)
print("Decimal:", decimal)

# Perform binary multiplication
binary, decimal = binary_multiplication(binary1, binary2)
print("Multiplication Result:")
print("Binary:", binary)
print("Decimal:", decimal)

# Perform binary division
binary, decimal = binary_division(binary1, binary2)
print("Division Result:")
print("Binary:", binary)
print("Decimal:", decimal)
```

```python
# Creating: Create a binary to decimal/decimal to binary converter
# Binary to decimal converter
def binary_to_decimal(binary):
    decimal = 0
    power = len(binary) - 1
    for digit in binary:
        decimal += int(digit) * 2**power
        power -= 1
    return decimal

# Decimal to binary converter
def decimal_to_binary(decimal):
    binary = ""
    while decimal > 0:
        remainder = decimal % 2
        binary = str(remainder) + binary
        decimal = decimal // 2
    return binary

# Take user input for binary or decimal numbers
input_type = input("Enter input type (B for binary, D for decimal): ").upper()

if input_type == "B":
    binary = input("Enter binary number: ")
    decimal = binary_to_decimal(binary)
    print("Decimal:", decimal)
elif input_type == "D":
    decimal = int(input("Enter decimal number: "))
    binary = decimal_to_binary(decimal)
    print("Binary:", binary)
else:
    print("Invalid input type.")
decimal = 43
binary = decimal_to_binary(decimal)
print("Decimal:", decimal)
print("Binary:", binary)
```

```python
### First Coding CHallenge
import csv

data = []
with open('/home/shivansh/devops/_notebooks/files/binary.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        for value in row:
            try:
                data.append(int(value))
            except ValueError:
                pass

# Calculate mean of all data
mean_all = sum(data) / len(data)

# Iterate over data to apply stipulations
processed_data = []
count_odd = 0
for value in data:
    if value % 2 == 1:
        count_odd += 1
        continue
    elif value % 5 == 0:
        processed_data.append(value * 2)
    else:
        processed_data.append(value)

# Calculate mean of processed data
mean_processed = sum(processed_data) / len(processed_data)

print(f"Mean of all data: {mean_all}")
print(f"Mean of processed data: {mean_processed}")
print(f"Number of odd values: {count_odd}")
```

2. What is the difference between boolean operations and logic gates?

Boolean operations are mathematical operations that involve Boolean values (True or False) and are used to determine the truth value of logical expressions. Logic gates, on the other hand, are physical or electronic devices that implement Boolean operations by taking one or more input signals and producing an output signal based on a specific logical function. In short, Boolean operations are abstract concepts used in logic and mathematics, while logic gates are the physical components that implement these concepts in digital circuits.
